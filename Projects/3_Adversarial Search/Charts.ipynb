```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import random
```

```python
bias_param_const = [0.00,0.25,0.50,0.75,1.00,1.25,1.50,1.75,2.00]
winning_percentage = [8.0,42.0,54.0,50.0,50.0,60.0,40.0,58.0,46.0]
plt.plot(bias_param_const, winning_percentage, color='g')
plt.xlabel('Bias Parameter')
plt.ylabel('Winning Percentage')
plt.title('Effect of bias param on winning percentage')
plt.show()
```

```python
raw_data = {'Opponent': ['MCTS', 'AlphaBeta'],
        'Random': [97.5, 83.0],
        'Greedy': [79.5, 71.5],
        'Minimax': [60.5, 46.0]}
df = pd.DataFrame(raw_data, columns = ['Opponent', 'Random', 'Greedy', 'Minimax'])
df
```

```python
n_groups = 3

winsMCTS = [97.5, 79.5, 60.5]
winsAlphaBeta = [83.0, 71.5, 46.0]


# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8
 
rects1 = plt.bar(index, winsMCTS,bar_width,
alpha=opacity,
color='g',
label='MCTS')
 
rects2 = plt.bar(index + bar_width, winsAlphaBeta, bar_width,
alpha=opacity,
color='r',
label='ABID')
 
plt.xlabel('Opponent')
plt.ylabel('Winning Perecentage')
plt.title('MCTS (Monte Carlo Tree Search) vs ABID (Alpha Beta Iterative Deepening)')
plt.xticks(index + bar_width, ('Random', 'Greedy', 'Minimax'))
plt.legend()
 
plt.tight_layout()
plt.show()
```
